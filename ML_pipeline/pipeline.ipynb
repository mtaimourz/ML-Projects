```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("titanic.csv")
```

- SimpleImputer b/c we have missing values
- OneHotEncoding b/c we have  categorical data
- MinMaxScalar for scaling
- DecisionTreeClassifier for prediction

```python
# remove columns that we do not need
df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"],inplace=True)
df.head()
```

## Train Test split

We need our model to predict if the person is going to live or not

We drop survived from x(input,independent) and take all others columns and in y then we put it again as output(dependent)

Why is train–test split important?

Without split:
 - Model memorizes answers 
 - Gives fake high accuracy
 - Fails on real data

With split:
 - Model learns patterns 
 - Honest performance check
 - Works in real life


```python
# test_size=0.2 → 20% data for testing, 80% data for training
# random_state=0 → same split every time (for consistency)

X_train, X_test, y_train, y_test  = train_test_split(df.drop(columns=['Survived']),
                                                     df['Survived'],
                                                     test_size=0.2,
                                                     random_state=42
)
```

```python
X_train.head(2)
```

```python
y_train.head(2)
```

```python
df.isnull().sum()
```

```python
# Apply imputation

si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy="most_frequent")

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

- Imputation means: Filling missing data (NaN values) instead of deleting rows.

- SimpleImputer() → tool from sklearn to fill missing values
 - Default strategy = mean


```python
X_train_age
```

```python
X_train_embarked
```

## One Hot Encoding

- Now we have two columns Age and Embarked they in both are categorical data. So we need to transform them into numerical data so use onehotencoding

- Computers work with numbers, not words. But columns like Sex and Embarked contain words (e.g., male/female, port names). So we must convert words into numbers.

- Why is handle_unknown=“ignore” used?
 - If a new value appears (something not seen before). The program won’t crash

```python
ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

x_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

x_test_sex = ohe_sex.fit_transform(X_test[['Sex']])
x_test_embarked = ohe_embarked.fit_transform(X_test_embarked)
```

```python
x_train_sex
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_rem
```

```python
X_train_transformed = np.concatenate([X_train_rem, X_train_age, x_train_embarked, x_train_sex], axis=1)
X_test_transformed = np.concatenate([X_test_rem, X_test_age, x_test_embarked, x_test_sex], axis=1)
```

```python
X_train_transformed
```

## Decision Tree Classifier

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Using ML Pipeline

```python
import numpy as np
import pandas as pd 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

In a pipeline, we first preprocess the data, then apply transformations like encoding and scaling, and finally train the model.
Pipeline ensures the same steps are applied to both training and testing data and prevents data leakage.

We have missing data in Age & Embarked columns

## Steps of Pipeline

- Train Test split 
- impute(fill) missing values by using one column transformer.
- (OHE) is output of step 1 send to second column transformer. Work of this column is to ohe of two columns(sex & embarked).
- scaling because all columns comes in similar numerical range.
- feature Selection (means we take 5 best columns(features)
- Decision Tree (we train model in this step by using decision tree that will give me output)

```python
df.drop(columns=["PassengerId","Name","Ticket","Cabin"],inplace=True)
# inplace means changes will be applied to original dataframe. 
# If we do inplace false a new dataframe will be created and original dataframe will remain same.
```

```python
df
```

```python
# Step 1: Test train split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Survived"]),
                                                   df["Survived"], test_size=0.2, random_state=42)

# random state is used to get same split every time we run the code. 42 is just a number we can use any integer.
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
#  Step 2 : Imputation transformer

trf1 = ColumnTransformer(
    [
        ("impute_age", SimpleImputer(), [2]), #  2 is index value of age column 
        # its a good strategy  to add index values instead of column names 
        # because in further steps if we do one hot encoding column names will change but index values will remain same
        ("impute_embarked", SimpleImputer(strategy="most_frequent"), [6]) # 6 is column name
    ],
    remainder="passthrough"
)

# passthrough means all other columns except the mentioned columns will remain the same
# most frequent means mode value will be used for imputation
```

column transformer is used to apply different transformations to different columns.

```python
# Step 3: # One hot encoding 
# converting categorical values into numerical values by creating new columns for each category.

trf2 = ColumnTransformer(
    [
        ("one_sex_embarked", OneHotEncoder(sparse_output=False, handle_unknown="ignore"), [1,6])
    ],
    remainder="passthrough"
    
)
```

```python
# Step 4: # Scaling
# converting values into a specific range. Here we are using min max scaler which converts values into range of 0 to 1.

trf3 = ColumnTransformer(
    [
        ("scale",MinMaxScaler(),slice(0,10)) # slice(4,12) means columns from index 4 to index 11
    ]
)

# Here we use minmiaz scaling not standard scaler because min max scaler keeps the distribution of data same
# whereas standard scaler changes the distribution of data. 

# In feature selection as we do we use minmax scaling because feature selection techniques are sensitive to distribution of data.

# we apply minmax scaling on all column if we want to apply on specific columns we can use index values instead of slice.

# We write 10 in slicing because after one hot encoding total 10 columns will be there.
```

```python
# Step 5: # Feature Selection

# Feature selection means selecting best features from the dataset.

trf4 = SelectKBest(score_func=chi2, k=8) # k is number of best features to select
```

```python
# Step 6: # Train the model
trf5 = DecisionTreeClassifier()
# decision tree classifier means we are using decision tree algorithm for classification.
```

```python
pipe = Pipeline([
    ("trf1", trf1),
    ("trf2", trf2),
    ("trf3", trf3),
    ("trf4", trf4),
    ("trf5", trf5)
])

# Pipeline is used to combine multiple steps into one. Here we are combining all the transformers and the classifier into one pipeline.
# here we pass list o f tuples where each tuple contains name of the step and the transformer or classifier object.

# Alternate syntax
# pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
pipe.fit(X_train, y_train)
```
