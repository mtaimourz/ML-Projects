They are related to Curse of dimensionality

Features(columns), You dont need to take very little features and not too much, take optimal features.

If features are too much then we use dimenationality reduction & in this we choose feature selection or feature extraction.

## Curse of Dimensionality

It means when the number of features (columns) increases, machine learning models become harder, slower, and less accurate.

More features ≠ better model

Happens mostly on images & text high dimension data

#### What Happens When Dimensions Increase?

1. Data Becomes Sparse
- Data points spread out
- Empty space increases
- Neighbors are far apart

2. Distance Becomes Meaningless
- Algorithms like KNN, K-Means depend on distance. In high dimensions, all points look equally far and Model gets confused

3. Need More Data
- More features → need exponentially more data Otherwise model overfits

4. Training Becomes Slow
- More computation and Higher memory usage



#### How to Handle Curse of Dimensionality?

1. Feature Selection
- Remove useless features, use correlation, importance

2. Feature Engineering
- Combine features, create meaningful features

3. Dimensionality Reduction
- PCA, LDA, Autoencoders, Regularization, L1 (Lasso), L2 (Ridge)

4. Use Robust Models
- Tree-based models, Boosting models
