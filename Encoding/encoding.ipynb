We remove the columns due to the concept knows as multi colinearity. There is a mathematical corelation b/w these columns.

They are dependent on each other. In ML, the input colums must be independent.

## 4. One Hot Encoding with Top Categories

This replaces brands with ≤100 occurrences as "uncommon" before encoding. 

```python
counts = df["brand"].value_counts()
counts
```

```python
df['brand'].nunique()
```

```python
threshold = 100 
repl = counts[counts <= threshold].index
```

```python
new = pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```
