# Multivariant Imputation Technique

In univariant we take data from the same column like mean/median and then fill missing column.

In Multivariant we take data from other multiple columns then fill missing values

There are many techniques for multivariant.

- KNN Imputer
- Iterativve Imputer

## KNN Imputer

KNN Imputer fills missing values by looking at the most similar data points and using their values.

When to Use KNN Imputer

- When data points are related
- When accuracy matters
- When missing values are not too many

 When NOT to Use

- Very large datasets (slow)
- Too many missing values
- Data with no similarity pattern

This technique is more accurate. If you have small data or mediam size data then you can work on it because it give better result as compare to mean/medain method.

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('../datasets/titanic/train.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
df.isnull().mean() * 100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
knn = KNNImputer(n_neighbors=3,weights='distance')

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf,y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```
